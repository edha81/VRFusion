```python
#簡單線性回歸 y =w*x+b
import pandas as pd
#url = 'https://raw.githubusercontent.com/GrandmaCan/ML/main/Resgression/Salary_Data.csv'
#data = pd.read_csv(url) 網路下載
data = pd.read_csv('Salary_Data.csv') #在相同資料夾

x = data['YearsExperience']
y = data['Salary']
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl #字體
from matplotlib.font_manager import fontManager
fontManager.addfont('ChineseFont.ttf') #新增字體
mpl.rc('font',family='ChineseFont') #使用字體


plt.scatter(x,y, marker='x',color='red') #marker 符號
plt.title('年資-薪水') #標題
plt.xlabel('年資')
plt.ylabel('月薪(千)')
plt.show()
```

```python
#試畫直線



def plot_pred(w,b):
    y_pred = x*w+b
    plt.plot(x,y_pred,color='blue',label='預測線')
    plt.scatter(x,y, marker='x',color='red',label='真實數據') #marker 符號
    plt.title('年資-薪水') #標題
    plt.xlabel('年資')
    plt.ylabel('月薪(千)')
    plt.xlim([0,12])#x軸範圍
    plt.ylim([0,140])#y軸範圍
    plt.legend()#顯示圖例 LABEL的部分
    plt.show()

plot_pred(10,20)
```

```python
from ipywidgets import interact#新增互動式元件
interact(plot_pred, w=(-100,100,1),b=(-100,100,1))#第一個為函數名稱，W跟B的大小值跟每次移動距離
```

最適合的值線
以資料1，2，3為例
w = 0 b = 0 時 (1-0)^2+(2-0)^2+(3-0)^2=14
w = 2 b = 0 時也一樣是14
而w = 1 b = 0時則為0，因此此為最適合之值線
